```python
import sqlalchemy as sal
from sqlalchemy.types import VARCHAR, NVARCHAR, INTEGER

# Define your exact schema here
# This prevents Python from "guessing" and using the wrong types
netflix_schema = {
    "show_id": VARCHAR(10),
    "type": VARCHAR(10),
    "title": NVARCHAR(200),        # NVARCHAR handles the '???'
    "director": VARCHAR(250),
    "cast": VARCHAR(1000),
    "country": VARCHAR(150),
    "date_added": VARCHAR(20),
    "release_year": INTEGER,
    "rating": VARCHAR(10),
    "duration": VARCHAR(10),
    "listed_in": VARCHAR(100),
    "description": VARCHAR(500)
}

# Now run your upload
with engine.connect() as conn:
    df.to_sql(
        'netflix_raw', 
        con=conn, 
        index=False, 
        if_exists='replace', 
        dtype=netflix_schema  # <--- This is the key!
    )
    print("✅ Table created with correct lengths and Unicode support")
```

```python
df.to_sql(
    'netflix_raw',
    con=conn,
    index=False,
    if_exists='append',
    dtype={
        'title': sal.types.NVARCHAR(length=200),
        'description': sal.types.NVARCHAR(length=1000)
    }
)
```
